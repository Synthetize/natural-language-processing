```python
import numpy as np
from gensim.models import KeyedVectors

# 1. Definisci le coppie di parole di genere (Gender Pairs)
# Queste coppie dovrebbero essere neutre in termini di stereotipi sociali
# (cioè, la differenza dovrebbe riguardare solo il genere).
# Una lista standard usata nella ricerca (es. Bolukbasi et al., 2016):
GENDER_PAIRS = [
    ('woman', 'man'),
    ('girl', 'boy'),
    ('she', 'he'),
    ('mother', 'father'),
    ('sister', 'brother'),
    ('aunt', 'uncle'),
    ('niece', 'nephew'),
    ('daughter', 'son'),
    ('female', 'male'),
    ('her', 'his')
]

def calculate_gender_direction_vector(embedding_model: KeyedVectors, pairs: list) -> np.ndarray:
    gender_difference_vectors = []

    # 2. Iterare sulle coppie e calcolare la differenza (V_femminile - V_maschile)
    for female_word, male_word in pairs:
        if female_word in embedding_model and male_word in embedding_model:
            # Calcola la differenza vettoriale: V_femminile - V_maschile
            diff_vector = embedding_model[female_word] - embedding_model[male_word]
            gender_difference_vectors.append(diff_vector)

    # 3. Calcolare il vettore medio (l'asse di genere non normalizzato)
    mean_gender_vector = np.mean(gender_difference_vectors, axis=0)
    print(mean_gender_vector)

    # 4. Normalizzare il vettore per avere lunghezza 1
    # Questo è importante per rendere i calcoli di proiezione (dot product/cosine similarity) comparabili
    normalized_gender_vector = mean_gender_vector / np.linalg.norm(mean_gender_vector)

    return normalized_gender_vector


gender_vector_twitter = calculate_gender_direction_vector(glove_twitter, GENDER_PAIRS)
gender_vector_wiki = calculate_gender_direction_vector(glove_wiki, GENDER_PAIRS)
```

```python
import numpy as np
from gensim.models import KeyedVectors
from typing import List, Dict

professions = ['doctor', 'nurse', 'engineer', 'teacher', 'programmer', 'scientist', 'soldier', 'pilot', 'dancer', 'driver', 'mechanic', 'lawyer', 'cleaner']

def calculate_projection_bias_scores(
    embedding_model: KeyedVectors,
    gender_vector: np.ndarray,
    professions: List[str]
) -> Dict[str, float]:
    bias_scores = {}
    normalized_gender_vector = gender_vector / np.linalg.norm(gender_vector)

    for profession in professions:
        if profession in embedding_model:
            # 1. Ottieni il vettore della professione (P)
            P_vector = embedding_model[profession]

            # 2. Calcola il Bias Score tramite il prodotto scalare (proiezione)
            # Nota: La somiglianza del coseno tra un vettore normalizzato e un altro vettore
            # (anche se non normalizzato) è il prodotto scalare.
            bias_score = np.dot(P_vector, normalized_gender_vector)

            bias_scores[profession] = bias_score

    return bias_scores


results_twitter = calculate_projection_bias_scores(glove_twitter, gender_vector_twitter, professions)
results_wiki = calculate_projection_bias_scores(glove_wiki, gender_vector_wiki, professions)
sorted_results_twitter= sorted(results_twitter.items(), key=lambda item: item[1], reverse=True)
sorted_results_wiki = sorted(results_wiki.items(), key=lambda item: item[1], reverse=True)

print("Bias Scores (Twitter):")
for profession, score in sorted_results_twitter:
    print(f"  {profession.capitalize():<12}: {score:.4f}")

print("\nBias Scores (Wikipedia):")
for profession, score in sorted_results_wiki:
    print(f"  {profession.capitalize():<12}: {score:.4f}")

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from typing import Dict, List

def plot_gender_bias_v2(
    bias_scores: Dict[str, float],
    title: str = "Gender Bias in Word Embeddings (Projection Score)",
    he_label: str = "← he",
    she_label: str = "she →"
):
    """
    Crea un grafico a barre di polarizzazione, risolvendo il DeprecationWarning di Seaborn.
    """

    # 1. Preparazione dei dati: Converti in DataFrame
    df = pd.DataFrame(list(bias_scores.items()), columns=['Word', 'Score'])

    # Crea la nuova colonna 'Gender Association' per il parametro hue
    df['Gender_Association'] = df['Score'].apply(
        lambda x: 'Female-Associated' if x >= 0 else 'Male-Associated'
    )

    # Ordina il DataFrame in base al punteggio
    df = df.sort_values(by='Score', ascending=False)

    # 2. Imposta lo stile del grafico
    sns.set_style("whitegrid")
    plt.figure(figsize=(8, len(df) * 0.45))

    # 3. Definisci una mappatura colori esplicita (blu per female, rosso per male)
    # Questa mappa sostituisce la tavolozza 'RdBu' definita automaticamente prima
    color_map = {
        'Female-Associated': sns.color_palette("Blues_d")[2],  # Tonalità di blu per 'she'
        'Male-Associated': sns.color_palette("Reds_d")[2]      # Tonalità di rosso/arancione per 'he'
    }

    # 4. Crea il grafico a barre usando 'hue' e impostando 'legend=False'
    ax = sns.barplot(
        x='Score',
        y='Word',
        data=df,
        hue='Gender_Association',      # ASSEGNA IL GRUPPO A HUE
        palette=color_map,             # USA LA TAVOLOZZA DEFINITA
        legend=False,                  # RIMUOVI LA LEGGENDA (come richiesto dal warning)
        dodge=False,                   # Assicura che le barre non siano affiancate
        edgecolor='gray'
    )

    # 5. Personalizzazione estetica
    ax.xaxis.grid(True)
    ax.yaxis.grid(False)
    ax.set_xlabel("Direction Projection", fontsize=12)
    ax.set_ylabel("Words", fontsize=12)
    ax.tick_params(axis='y', labelsize=10)

    # Inserisce le etichette "he" e "she" nella parte superiore
    plt.title(title, fontsize=14, pad=20)
    ax.text(ax.get_xlim()[0] * 0.9, ax.get_ylim()[1] + 1, he_label,
            ha='left', va='bottom', fontsize=12)
    ax.text(ax.get_xlim()[1] * 0.9, ax.get_ylim()[1] + 1, she_label,
            ha='right', va='bottom', fontsize=12)

    plt.tight_layout()
    plt.show()
plot_gender_bias_v2(results_twitter)
plot_gender_bias_v2(results_wiki)
```

```python
import numpy as np
from gensim.models import KeyedVectors
from typing import List, Dict

def neutralize_vector(professions_vector: np.ndarray, gender_vector: np.ndarray) -> np.ndarray:
    projection_coeff = np.dot(professions_vector, gender_vector)
    return professions_vector - projection_coeff * gender_vector

def apply_debias(
    embedding_model,
    gender_vector: np.ndarray,
    professions: List[str]
) -> any:

    #normalized_gender_vector = gender_vector / np.linalg.norm(gender_vector)

    for profession in professions:
        if profession in embedding_model:
            embedding_model[profession] = neutralize_vector(
                embedding_model[profession],
                gender_vector
            )

    return embedding_model

def print_debias_results(
    debiased_model,
    gender_vector: np.ndarray,
    professions: List[str]
):
    bias_scores: Dict[str, float] = {}

    normalized_gender_vector = gender_vector / np.linalg.norm(gender_vector)

    for profession in professions:
        if profession in debiased_model:
            score = np.dot(debiased_model[profession], normalized_gender_vector)
            bias_scores[profession] = score

    print("Professions  | Score")
    sorted_scores = sorted(bias_scores.items(), key=lambda item: abs(item[1]), reverse=True)

    for profession, score in sorted_scores:
        print(f"{profession.capitalize():<12} | {score: .8f}")
    print("-" * 35)


# --- Esempio di Utilizzo ---


# Assumi che 'glove_twitter' e 'gender_vector_twitter' esistano

debiased_glove_twitter = apply_debias(
    glove_twitter,
    gender_vector_twitter,
    professions
)

print_debias_results(
    debiased_glove_twitter,
    gender_vector_twitter,
    professions
)

debiased_glove_wiki = apply_debias(
    glove_wiki,
    gender_vector_wiki,
    professions
)

print_debias_results(
    debiased_glove_wiki,
    gender_vector_wiki,
    professions
)
```

```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt


def run_pca_on_professions(embedding_model: KeyedVectors, professions: List[str], n_components: int = 2):
    # Collect embeddings
    vectors = []
    valid_professions = []
    for profession in professions:
        if profession in embedding_model:
            vectors.append(embedding_model[profession])
            valid_professions.append(profession)

    vectors = np.array(vectors)

    # Run PCA
    pca = PCA(n_components=n_components)
    reduced = pca.fit_transform(vectors)

    return reduced, valid_professions, pca


# Run PCA for Twitter and Wiki embeddings
pca_twitter, valid_prof_twitter, model_twitter = run_pca_on_professions(glove_twitter, professions)
pca_wiki, valid_prof_wiki, model_wiki = run_pca_on_professions(glove_wiki, professions)


# Plot PCA results
def plot_pca(pca_result, labels, title):
    plt.figure(figsize=(8, 6))
    x = pca_result[:, 0]
    y = pca_result[:, 1]
    plt.scatter(x, y, alpha=0.7)

    for i, label in enumerate(labels):
        plt.text(x[i] + 0.01, y[i] + 0.01, label, fontsize=9)

    plt.xlabel("PC1")
    plt.ylabel("PC2")
    plt.title(title)
    plt.grid(True)
    plt.show()


plot_pca(pca_twitter, valid_prof_twitter, "PCA of Professions - Twitter Embeddings")
plot_pca(pca_wiki, valid_prof_wiki, "PCA of Professions - Wikipedia Embeddings")


# Compare PCA direction with gender vector
def compare_pca_with_gender(pca_model, gender_vector, n_components=2):
    # Project gender vector onto principal components
    components = pca_model.components_[:n_components]  # top PCs
    gender_projection = [np.dot(gender_vector, pc) for pc in components]
    explained_variance = pca_model.explained_variance_ratio_[:n_components]

    for i, (proj, var) in enumerate(zip(gender_projection, explained_variance), start=1):
        print(f"PC{i} alignment with gender vector: {proj:.4f}, explains {var * 100:.2f}% variance")


print("\nTwitter Embeddings PCA vs Gender Vector:")
compare_pca_with_gender(model_twitter, gender_vector_twitter)

print("\nWikipedia Embeddings PCA vs Gender Vector:")
compare_pca_with_gender(model_wiki, gender_vector_wiki)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from gensim.models import KeyedVectors

# --- Carica i modelli salvati ---
glove_wiki = KeyedVectors.load("glove-wiki-gigaword-100")
glove_twitter = KeyedVectors.load("glove-twitter-100")

# --- Coppie razziali ---
RACIAL_PAIRS = [
    ("black", "white"),
    ("african", "european"),
    ("african_american", "caucasian"),
    ("slave", "master"),
    ("ghetto", "suburb"),
]

# --- Parole target (professioni + tratti) ---
TARGET_WORDS = [
    "doctor", "nurse", "engineer", "lawyer",
    "criminal", "intelligent", "dangerous",
    "police", "gang", "scientist", "teacher"
]


def calculate_racial_direction(embedding_model, pairs):
    """Calcola il vettore di direzione razziale"""
    diffs = []
    for w1, w2 in pairs:
        if w1 in embedding_model and w2 in embedding_model:
            diffs.append(embedding_model[w1] - embedding_model[w2])
    mean_vector = np.mean(diffs, axis=0)
    return mean_vector / np.linalg.norm(mean_vector)


def run_pca(embedding_model, words):
    """Esegue PCA su una lista di parole"""
    vectors = []
    valid_words = []
    for word in words:
        if word in embedding_model:
            vectors.append(embedding_model[word])
            valid_words.append(word)

    vectors = np.array(vectors)
    pca = PCA(n_components=2)
    reduced = pca.fit_transform(vectors)
    return reduced, valid_words, pca


def plot_pca(reduced, words, title):
    """Plot PCA scatter con etichette"""
    plt.figure(figsize=(7, 6))
    x = reduced[:, 0]
    y = reduced[:, 1]

    plt.scatter(x, y, alpha=0.7)
    for i, word in enumerate(words):
        plt.text(x[i] + 0.02, y[i] + 0.02, word, fontsize=10)

    plt.xlabel("PC1")
    plt.ylabel("PC2")
    plt.title(title)
    plt.grid(True)
    plt.show()


def check_pca_racial_alignment(pca_model, racial_vector, title):
    """Verifica quanto PC1 e PC2 si allineano con il vettore razziale"""
    components = pca_model.components_[:2]
    racial_proj = [np.dot(racial_vector, pc) for pc in components]
    explained = pca_model.explained_variance_ratio_[:2]

    print(f"\n{title}")
    for i, (proj, var) in enumerate(zip(racial_proj, explained), start=1):
        print(f"PC{i} alignment with racial direction: {proj:.4f} (explains {var * 100:.2f}% variance)")


# --- Twitter embeddings ---
racial_vector_twitter = calculate_racial_direction(glove_twitter, RACIAL_PAIRS)
reduced_twitter, valid_words_twitter, pca_twitter = run_pca(glove_twitter, TARGET_WORDS)
plot_pca(reduced_twitter, valid_words_twitter, "PCA: Twitter Racial Bias")
check_pca_racial_alignment(pca_twitter, racial_vector_twitter, "Twitter Embeddings")

# --- Wikipedia embeddings ---
racial_vector_wiki = calculate_racial_direction(glove_wiki, RACIAL_PAIRS)
reduced_wiki, valid_words_wiki, pca_wiki = run_pca(glove_wiki, TARGET_WORDS)
plot_pca(reduced_wiki, valid_words_wiki, "PCA: Wikipedia Racial Bias")
check_pca_racial_alignment(pca_wiki, racial_vector_wiki, "Wikipedia Embeddings")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Pick the pair you want
GENDER_PAIR = ("he", "she")  # or ("she", "he")

professions = ['doctor', 'nurse', 'engineer', 'teacher',
               'programmer', 'scientist', 'soldier', 'pilot',
               'dancer', 'driver', 'mechanic', 'lawyer', 'cleaner']

def plot_pca_one_pair(embedding_model, gender_pair, professions):
    words = [gender_pair[0], gender_pair[1]] + professions
    words = [w for w in words if w in embedding_model]

    X = np.array([embedding_model[w] for w in words])
    labels = words

    # PCA
    pca = PCA(n_components=2)
    X_pca = pca.fit_transform(X)

    # Plot
    plt.figure(figsize=(10, 8))
    for i, word in enumerate(labels):
        if word in professions:
            color = "orange"
            marker = "o"
        else:  # gender pair words
            color = "blue"
            marker = "s"
        plt.scatter(X_pca[i, 0], X_pca[i, 1], c=color, s=100, alpha=0.8, marker=marker, edgecolor="k")
        plt.text(X_pca[i, 0] + 0.05, X_pca[i, 1] + 0.05, word, fontsize=9)

    plt.title(f"PCA: {gender_pair[0]} vs {gender_pair[1]} + professions")
    plt.xlabel("PC1")
    plt.ylabel("PC2")
    plt.grid(True, linestyle="--", alpha=0.4)
    plt.show()

# Example usage:
plot_pca_one_pair(glove_wiki, GENDER_PAIR, professions)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

professions = ['doctor', 'nurse', 'engineer', 'teacher',
               'programmer', 'scientist', 'soldier', 'pilot',
               'dancer', 'driver', 'mechanic', 'lawyer', 'cleaner']

def plot_pca_with_gender_axis(embedding_model, gender_vector, professions, ref_words=("man", "woman")):
    # Collect words (professions + reference words)
    words = [w for w in professions + list(ref_words) if w in embedding_model]

    X = np.array([embedding_model[w] for w in words])
    labels = words

    # PCA
    pca = PCA(n_components=2)
    X_pca = pca.fit_transform(X)

    # Project gender vector into PCA space
    gender_axis = pca.transform([gender_vector])[0]

    # Plot
    plt.figure(figsize=(10, 8))
    for i, word in enumerate(labels):
        if word in professions:
            color = "orange"
            marker = "o"
        else:  # ref words man/woman
            color = "blue"
            marker = "s"
        plt.scatter(X_pca[i, 0], X_pca[i, 1], c=color, s=100, alpha=0.8, marker=marker, edgecolor="k")
        plt.text(X_pca[i, 0] + 0.05, X_pca[i, 1] + 0.05, word, fontsize=9)

    # Draw gender axis arrow
    plt.arrow(0, 0, gender_axis[0], gender_axis[1],
              color="red", width=0.01, head_width=0.1, alpha=0.7, length_includes_head=True)
    plt.text(gender_axis[0], gender_axis[1], "Gender axis", color="red", fontsize=10)

    plt.title("PCA: Professions + Gender Reference with Gender Axis")
    plt.xlabel("PC1")
    plt.ylabel("PC2")
    plt.grid(True, linestyle="--", alpha=0.4)
    plt.show()

# Example usage with your computed gender vector for Twitter:
plot_pca_with_gender_axis(glove_wiki, gender_vector_twitter, professions)
```
